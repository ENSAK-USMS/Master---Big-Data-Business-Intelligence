### <center> TRAVAUX PRATIQUES : OPTIMISATION SANS CONTRAINTES


```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D
```

#### Exercice 1

Soient f,g les fonctions definies sur $R^2$ par : $$f(x,y) = x^3 + y^3 -9xy +27 \\ g(x,y)=x^4 + y^4 - (x+y)^2$$



```python
def showmin(f, x0, title='', step=100, rounding=None, elev=30, azim=30):
    result = minimize(f, x0)
    result.x = tuple(round(x, rounding) for x in result.x)
    result.fun = round(result.fun, rounding)
    x = np.linspace(-10, 10, step)
    y = np.linspace(-10, 10, step)
    X, Y = np.meshgrid(x, y)
    Z = f([X, Y, np.zeros_like(X)])
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_title(f'{title}\n Minimum at {result.x} with value {result.fun}')
    ax.view_init(elev=elev, azim=azim)
    plt.show()
```

Le gradient et la matrice hessienne de f valent :
$$\nabla f(x,y) = \begin{pmatrix}3x^2 - 9y \\ -9x + 3y^2 \end{pmatrix} ,  \nabla^2f = \begin{pmatrix}6x^2 - 9 \\ -9 + 6y \end{pmatrix}$$
Les points critiques de f sont les points O(0,0) et A(3,3).  
A(3,3) est lâ€™unique minimum local de f sur $R^2$


```python
# DÃ©finir la fonction
f = lambda x: x[0]**3 + x[1]**3 - 9*x[0]*x[1] + 27
# Choisir un point de dÃ©part initial pour l'optimisation
x0 = [1, 1]
title = 'f(x,y) = x^3 + y^3 - 9xy + 27'

showmin(f, x0,title)
```

Le gradient et la matrice hessienne de f valent :
$$\nabla g(x,y) = \begin{pmatrix}4x^3 - 2x -2y \\ 4y^3 - 2x -2y \end{pmatrix} ,  \nabla^2g = \begin{pmatrix}12x^2 - 2 & -2 \\ -2 & 12y^2 -2 \end{pmatrix}$$
Les points critiques de g sont les points O(0,0), A(1,1) et B(âˆ’1,âˆ’1)

A(1,1) et B(âˆ’1,âˆ’1) sont les deux minima global g(1,1) = g(-1,-1) = -2


```python

g = lambda x: x[0]**4 + x[1]**4 - (x[0]+x[1])**2 
x0 = [-4, -5]
title = 'g(x,y) = x^4 + y^4 - (x+y)^2'

showmin(g,x0,title)
```

#### Exercice 2

f est la fonction definie sur $R^3$
$$ f(x_1,x_2,x_3) = x_1^2 + x_2^2 + x^2_3 - x_1x_2 - x_2x_3 + 3x_1 -x_2 + 2x_3$$
La matrice hessienne H de f est :

$$ H = \begin{pmatrix}2 & -1 & 0 \\ -1 & 2 & -1 \\ 0 & -1 & 2 \end{pmatrix} $$

Elle est definie positive et la fonction f est donc strictement
convexe et admet un unique minimum global u solution du systeme lineaire H.u = âˆ’b ou $b^t$ = (3,âˆ’1,2)

le minimum trouver est g(-2.25,-1.5,-1.75) = -4.375


```python
f  = lambda x: x[0]**2 + x[1]**2 + x[2]**2 - x[0]*x[1] -x[1]*x[2] + 3*x[0] - x[1] + 2*x[2]
x0 = [0, 0, 0]
title = 'f(x,y,z) = x^2 + y^2 + z^2 - xy - yz + 3x - y + 2z'
showmin(f, x0,title, rounding=2)
```

#### Exercice 3

soit $$ f_t(x,y)= \frac{t}{2}(x^2+y^2) + xy + x - y$$

La matrice Hessienne de $f_t$ est :
$$ H = \begin{pmatrix}t & 1\\1 & t\end{pmatrix} $$
elle a pour determinant $t^2 âˆ’1$

pour t = 0 elle est definie positive: $f_t$ a un unique minimum



```python
f = lambda x: t/2 * (x[0]**2 + x[1]**2) - x[0]*x[1] + x[0] - x[1]
x0 = [0, 0]
t = 0
title = 'f(x,y) = t/2(x^2+y^2) + x - y et t = 0'
showmin(f, x0, title ,elev=20,azim=150)
```

pour t = -1 elle est semi-definie negative


```python
t = -1
title = 'f(x,y) = t/2(x^2+y^2) + x - y et t = -1'
showmin(f, x0,title,elev=20,azim=150)
```

pour t = -2 elle est definie negative : $f_t$ a un unique maximum


```python
t = -2
title = 'f(x,y) = t/2(x^2+y^2) + x - y et t = -2'
showmin(f, x0,title,elev=20,azim=70)
```
