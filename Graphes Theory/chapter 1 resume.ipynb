## Introduction

### What is a graph?



A graph is a mathematical structure that consists of a set of vertices (also called nodes) and a set of edges (also called links or lines), which are connections between the vertices. The edges may be directed or undirected. An edge between two vertices is called an adjacency. A graph may be weighted (the edges have a weight) or unweighted.

Example of a graph:

```
Vertices: 0, 1, 2, 3, 4
Edges: (0, 1), (0, 2), (1, 2), (2, 3), (3, 4)
```

## Graphes Hamiltoniens et Euleriens

### Definitions

- `Graphe Hamiltonien`

Un graphe est dit hamiltonien s'il contient un cycle passant par chaque sommet exactement une fois.

- `Graphe Eulerien`

Un graphe est dit eulerien s'il contient un cycle passant par chaque arête exactement une fois.

Remarques: 

Un graphe hamiltonien est eulerien, mais l'inverse n'est pas vrai.

Si  le degre est >= n/2, alors le graphe est hamiltonien.

### Definition



Soit un graphe G = (V, E) avec n sommets et m arêtes. La matrice d'adjacence A de G est une matrice carrée de taille n x n, où n est le nombre de sommets du graphe. Les éléments de la matrice sont définis comme suit:


### example python

```python
def adjacency_matrix(graph):
    n = len(graph)
    adj_matrix = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in graph[i]:
            adj_matrix[i][j] = 1
    return adj_matrix
```

````python
def adjacency_matrix(graph):
    n = len(graph)
    adj_matrix = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in graph[i]:
            adj_matrix[i][j] = 1
    return adj_matrix

# use the function
graph = [[1, 2], [0, 2], [0, 1, 3], [2, 4, 5], [3, 5], [3, 4]]

# print it as a matrix
for row in adjacency_matrix(graph):
    print(' '.join(map(str, row)))
````
