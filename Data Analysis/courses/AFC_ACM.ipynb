# **(AFC)** 

> # définition
 L’Analyse Factorielle des Correspondances (AFC) a été développé par  Jean-Paul Benzécri durant la période 1970-1990,c'est une méthode factorielle de Statistique Descriptive Multidimensionnelle (MDS).Elle est principalement utilisée dans le domaine de l'analyse des données pour explorer des tableaux de contingence, c'est-à-dire des tableaux qui présentent des fréquences ou des pourcentages pour différentes catégories dans deux ou plusieurs variables catégorielles.
L'AFC est particulièrement utile lorsqu'on travaille avec des données catégorielles, telles que des enquêtes comportant des questions à choix multiples, des données de sondage, des données de marketing, etc. Elle permet de mettre en évidence les associations et les tendances dans les données catégorielles d'une manière plus compréhensible et interprétable.


> # Objectif de l'AFC
   L’analyse factorielle des correspondances vise à rassembler en un nombre réduit de
dimensions la plus grande partie de l’information initiale en s’attachant non pas aux valeurs
absolues mais aux correspondances entre les variables, c’est-à-dire aux valeurs relatives. Cette
réduction est d’autant plus utile que le nombre de dimensions initial est élevé. La notion de
“ réduction ” est commune à toutes les techniques factorielles – c’est-à-dire où l’on extrait des
facteurs – l’AFC offre la particularité (contrairement aux ACP) de fournir un espace de
représentation commun aux variables et aux individus. Pour cela l’AFC raisonne à partir de
tableau réduit ou de fréquences.
 L'objectif principal de l'AFC est de réduire la complexité des données tout en préservant l'essentiel de l'information contenue dans le tableau croisé des fréquences des catégories.

Voici les objectifs spécifiques de l'AFC :

### Réduction de la Dimensionnalité :
 L'AFC réduit la dimensionnalité des données en transformant un grand nombre de variables catégorielles en un espace de dimension plus faible. Cela facilite la visualisation et l'interprétation des données.

### Visualisation des Relations : 
L'objectif principal de l'AFC est de visualiser les relations entre les lignes et les colonnes du tableau de contingence. Elle permet de détecter des associations entre les différentes catégories des variables .


### Interprétation des Axes : 
L'interprétation des axes principaux de l'AFC permet de comprendre quelles catégories sont les plus influentes sur chaque axe, aidant ainsi à interpréter les relations dans les données.

### Analyse de Similarités : 
L'AFC peut être utilisée pour mesurer les similarités entre les différentes catégories des variables. Elle permet de voir quels groupes de catégories sont similaires ou dissimilaires, Cela peut être particulièrement utile dans des domaines tels que le marketing, où l'analyse des préférences des consommateurs etc.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fanalysis.ca import CA
import scipy
```

```python
D = pd.read_excel(r'C:\Users\DELL\Documents\AFC.xlsx',index_col=0)
print(D)
```

```python
tot_lig = np.sum(D.values,axis=1)
print(tot_lig)
tot_col = np.sum(D.values,axis=0)
print(tot_col)


```

```python
prof_lig = np.apply_along_axis(arr=D.values, axis=1 ,func1d=lambda x : x/np.sum(x))
print(prof_lig)
```

```python
somme = np.zeros(shape=(prof_lig.shape[0]))
for i in range(prof_lig.shape[1]):
  plt.barh(range(prof_lig.shape[0]),prof_lig[:,i],left=somme)
  somme = somme + prof_lig[:,i]

plt.yticks(range(prof_lig.shape[0]),D.index)
plt.show()
```

```python
prof_marg_lig = tot_col/np.sum(tot_col) #f.j
print(prof_marg_lig)
```

```python
distPairesLig = np.zeros(shape=(prof_lig.shape[0], prof_lig.shape[0]))

# double boucle
for i in range(prof_lig.shape[0] - 1):
    for j in range(prof_lig.shape[0]):
        distPairesLig[i, j] = np.sum((prof_lig[i, :] - prof_lig[j, :]) ** 2 / prof_marg_lig)

# affichage
print(pd.DataFrame(distPairesLig, index=D.index, columns=D.index))
```

```python
distoLig = np.apply_along_axis(arr=prof_lig,axis=1,func1d=lambda x:np.sum((x-prof_marg_lig)**2/prof_marg_lig))
#affichage
print(pd.DataFrame(distoLig,index=D.index))
```

```python
poidsLig = tot_lig/np.sum(tot_lig)
inertieLig = distoLig * poidsLig
#affichage
print(pd.DataFrame(np.transpose([distoLig,poidsLig,inertieLig]),columns=['Disto2','Poids','Inertie'] ,index=D.index))
#total inertie
tot_InertieLig = np.sum(inertieLig)
print("\ntot_InertieLig :", tot_InertieLig)
```

```python
afc = CA(row_labels=D.index,col_labels=D.columns)
afc.fit(D.values)
#information restituée sur les facteurs
print(afc.eig_)
print(pd.DataFrame(afc.row_coord_,index=D.index))
```

```python
#affichage graphique des v.p.
afc.plot_eigenvalues()
```

```python
#affichage dans le premier plan factoriel
fig, ax = plt.subplots(figsize=(10,10))
ax.axis([-0.5,+0.5,-0.5,+0.5])
ax.plot([-0.5,+0.5],[0,0],color='red',linestyle='--')
ax.plot([0,0],[-0.5,+0.5],color='red',linestyle='--')
ax.set_xlabel("Dim.1 (97.35% )")
ax.set_ylabel("Dim.2 (2.01%)")
plt.title("Carte des modalités lignes")
for i in range(D.shape[0]):
 ax.text(afc.row_coord_[i,0],afc.row_coord_[i,1],D.index[i])
plt.show()
```

```python
#tableau des profils colonnes
prof_col = np.apply_along_axis(arr=D.values,axis=0,func1d=lambda x:x/np.sum(x))
print(pd.DataFrame(prof_col,index=D.index,columns=D.columns))
```

```python
#profil marginal des filières# fi.
prof_marg_col = tot_lig/np.sum(tot_lig)
print(prof_marg_col)
```

```python
#distance**2 à l'orgine
distoCol = np.apply_along_axis(arr=prof_col,axis=0,func1d=lambda x:np.sum((x
-prof_marg_col)**2/prof_marg_col))
#affichage
print(pd.DataFrame(distoCol,index=D.columns))
```

# **(ACM)**


> ## définition

L’**ACM** est une généralisation de l’AFC pour $p>2$ variables qualitatives. Elle permet d’analyser les liens entre $p$ variables qualitatives. 
L’ACM est une AFC faite sur le **tableau disjonctif** complet (ou le tableau de Burt). Le poids de chacune des variables est donc le nombre de ses modalités moins 1. Cela signifie qu’une variable avec un grand nombre de modalités pèsera plus dans la constitution des axes qu’une autre avec peu de modalités. 

- Remarque:On peut y mettre des variable quantitatives, à condition de les discrétiser (donc de les transformer en variables qualitatives).

> ## Objectif de l'ACM

l'objectif principal de l'ACM est de permettre une analyse intégrée de plusieurs tableaux de données catégorielles afin d'explorer les relations complexes et hétérogènes entre différentes sources de données, tout en fournissant des outils visuels pour faciliter la compréhension et l'interprétation des résultats.
Ainsi que l'identification des Associations et des Structures, la visualisation des Données ,l'aider à la Prise de Décision ,etc...

> ## PRINCIPE DE L’ANALYSE DES CORRESPONDANCES MULTIPLES

- *Calcul des Tableaux de Contingence*: Créer des tableaux de contingence pour chaque ensemble de données catégorielles.


|        | Identifiant|  sexe    |Age    |
|--------|-----------|-----------| ----  | 
| 1      |0001       | H         | <20   |
| 2      |0002       | F         |  20-60|
| 3      |  0003     | F         | >60   |
| 4      | 0004      | H         |  <20  |
- *Le Tableau Disjonctif Complet (TDC)*: est un tableau de contingence spécifique utilisé dans l'Analyse des Correspondances Multiples (ACM) et d'autres techniques d'analyse factorielle pour les données catégorielles.
Le TDC est appelé "disjonctif complet" parce qu'il prend en compte toutes les combinaisons possibles de catégories des différentes variables.
La construction du tableau disjonctif complet est l'une des étapes préalables au calcul de l’Analyse des Correspondances Multiples. Les p variables qualitatives sont éclatées en p tableaux disjonctifs Z1, Z2, …, Zp, composés d'autant de colonnes qu'il y a de modalités pour chacune des variables. A chaque fois qu'une modalité m de la jème variable correspond à un individu i, on affecte 1 à Zj(i,m). Les autres valeurs de Zj sont nulles. Les p tableaux disjonctifs sont alors concaténés en un tableau disjonctif complet.

 --> On transforme les modalités des variables qualitatives en autant de variables binaires.
 




- *Calcul des Correspondances*: Effectuer des calculs basés sur le Le Tableau Disjonctif Complet pour obtenir les valeurs propres, les vecteurs propres, et les coordonnées des individus et des variables dans l'espace factoriel.
- *Analyse des Valeurs Propres*: Évaluer les valeurs propres pour déterminer le nombre optimal de dimensions à retenir.
Le traitement mathématique du tableau $X$ commence par le calcul de la matrice $Z=X/(IK)$, s'ensuit le calcul du vecteur $r$, qui contient la somme en ligne de la matrice Z et enfin, le calcul du vecteur $c$, qui contient la somme en colonne de la matrice $Z$.
On prend également en compte les matrices diagonales $D_{r}={diag}(r)$ et $D_{c}=diag(c)$, issues de r et c respectivement. 
> Le processus est similaire à celui de l'Analyse Factorielle des Correspondances :  

-on calcule la matrice M.

$M=D_{r}^{-1/2}(Z-rc^{t})D_{c}^{-1/2}$

-On calcule la matrice $S=M^{t}M$, puis on calcule les valeurs et vecteurs propres(us) de cette matrice. Les valeurs propres(λs) correspondent à l'inertie de chaque facteur. Les coordonnées des individus dans ce nouvel espace vectoriel sont données par la formule suivante. 
 
   $Fs=Mus$ 
 
La i-ième ligne de F contient les coordonnées du i-ième individu dans l'espace factoriel, tandis que les coordonnées des variables dans le même espace factoriel sont données par :

  $Gs=\sqrt{λs}*us$
- *le nombre maximal de facteurs $Hmax$*: $Hmax$ représente le nombre maximal de facteurs que vous pouvez extraire à partir des données. 
$M$ représente le nombre total de modalitées et $P$ représente le nombre total des variables. 
La formule pour calculer $Hmax$ est généralement : $Hmax=M−P$


- *Construction des Bi-Plots*: Créer des bi-plots pour visualiser les relations entre les variables et les individus dans un espace réduit.

- *Interprétation des Résultats*: Examiner les bi-plots pour interpréter les associations entre les variables et les individus. Identifier les 
        groupes de variables ou d'individus similaires, et détecter les tendances et les structures dans les données.

```python
# Pour gerer et manipuler les structures de données multidimentionnels
import pandas as pd  

# Pour gerer les calculs matriciels
import numpy as np  

# Pour visualiser les données
import matplotlib.pyplot as plt 
from fanalysis.mca import MCA

#mt est une instance de la classe math
import math as mt 
data = pd.read_excel(r'C:/Users/DELL/Documents/Data_Methodes_Factorielles.xlsx',sheet_name="ACM_CANINES",index_col=0)
print(data)
```

```python
DActives = [x for x in data.columns if data[x].dtype == 'object']
# On récupere les colonnes du dataSet.
variable = data[DActives]
print(variable)
# Le nombre des individus
n = variable.shape[0]
# Le nombre des variables
p = variable.shape[1]
print("le nombre des individus :",n)
print("le nombre des variables :",p)
# pd.get_dummies() est utilisé pour la manipulation des données. Il convertit les données catégorielles en variables 
# fictives ou indicatrices.  
# Construire le tableau disjonctif complet
TDC = pd.get_dummies(variable,prefix='',prefix_sep='')
print(TDC)
 # Le nombre total de modalité
M = TDC.shape[1]
print("Le nombre total de modalité",M)
#instanciation de la méthode mca qu'est disponible dans fanalysis
acm = MCA(row_labels=TDC.index,var_labels=TDC.columns) 
x=acm.fit(TDC.values)  
# nombre max de facteur (pour savoir la dimension de nouvelle espace)
#le nombre de facteurs extraits ne doit pas dépasser le nombre d'individus moins le nombre de variables. Cela garantit que vous ne créez pas plus de facteurs que le nombre d'observations que vous avez dans votre jeu de données
Hmax = M - p  
print("nombre max de facteur",Hmax)
# Détérmination des valeurs propres 
valeur_propre = pd.DataFrame(np.transpose(acm.eig_),columns = ['valeur propre','%','cumul %'])
print("les val propres :")
print(valeur_propre)
```

```python
# Eboulis des valeurs propres pour savoir les valeurs propres qui porte plus informations
#Son rôle est de vous aider à déterminer combien de composantes principales (ou facteurs) vous devez retenir pour réduire la dimensionnalité de vos données.
fix , ax = plt.subplots(figsize=(5,5)) 
ax.plot(range(0,len(valeur_propre)),acm.eig_[0],".-")  
ax.set_xlabel("nombre de facteurs")
ax.set_ylabel("valeur propre")
plt.title("Eboulis des valeus propres")
#seuil - Règle de kaiser
print(p)
ax.plot([1,Hmax],[1/p,1/p],"r-",linewidth=1) 
plt.show()

# Séléction les valeurs propres qui porte plus informations
valeur_qui_porte_plus_information = [acm.eig_[0][0] , acm.eig_[0][1]]
print("les valeurs propres qui portent plus d'information")
print(valeur_qui_porte_plus_information)
 # Affichage les coordonnes des facteurs.
data_facteur = pd.DataFrame(acm.col_coord_[0:len(TDC.columns),:2],columns=['F1','F2'],index=TDC.columns) 
data_facteurrow=pd.DataFrame(acm.row_coord_[:,:2],index=TDC.index,columns=['Coord.F1','Coord.F2'])
print(data_facteur)
print(data_facteurrow)
    # Pour détermine l'échelle correspondant 
maxii_F1 = max(list(data_facteur['F1'])) #pour recuperer la valeur maximale dans la 1ere facteur F1
minii_F1 = min(list(data_facteur['F1']))  #pour recuperer la valeur minimale dans la 1ere facteur F1
maxii_F2 = max(list(data_facteur['F2']))  #pour recuperer la valeur maximale dans le 2eme facteur F2
minii_F2 = min(list(data_facteur['F2']))  #pour recuperer la valeur minimale dans le 2eme facteur F2

if (maxii_F1>0):
        maxi_F1 = mt.floor(maxii_F1)+1
else:
        maxi_F1 = mt.floor(maxii_F1)

if (minii_F1<0): 
        mini_F1 = mt.floor(minii_F1)
else:
        mini_F1 =mt.floor(minii_F1)-1

if (maxii_F2>0): 
        maxi_F2 = mt.floor(maxii_F2)+1
else:
        maxi_F2 = mt.floor(maxii_F2)

if (minii_F2<0): 
        mini_F2 = mt.floor(minii_F2)
else:
        mini_F2 =mt.floor(minii_F2)-1
        
        
```
